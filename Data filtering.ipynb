```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from datetime import date
import datetime as dt

data = pd.read_csv('Data 2019.csv')
data['Sdate'] = pd.to_datetime(data['Sdate'], format='%d/%m/%Y %H:%M')
data.head()
```

```python
east_bound = data[(data['LaneDescription'] == 'East Bound')
                  & (data['Sdate'] < pd.to_datetime("2019/02/01"))
                  & (data['Sdate'] > pd.to_datetime("2019/01/01"))][['Sdate', 'Volume']]


west_bound = data[data['LaneDescription'] == 'West Bound'][['Sdate', 'Volume']]


byday = east_bound.groupby(pd.to_datetime(east_bound.Sdate).dt.date).agg({'Volume': 'sum'}).reset_index() 
byday['Sdate'] = byday['Sdate'].apply(lambda x: dt.datetime.strftime(x, '%Y-%m-%d'))
byday['Sdate'].tolist()
byday['Volume'].tolist()
```
